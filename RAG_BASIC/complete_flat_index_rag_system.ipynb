```python
import os
from dotenv import load_dotenv
from uuid import uuid4
from pypdf import PdfReader
from langchain.embeddings.huggingface import HuggingFaceInferenceAPIEmbeddings
from langchain.text_splitter import CharacterTextSplitter
from langchain.vectorstores import FAISS
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
loader = PyPDFLoader("/content/UNIT 5.pdf")
docs = loader.load()


splitter = RecursiveCharacterTextSplitter(chunk_size = 1000,chunk_overlap = 200)
splitter_docs = splitter.split_documents(docs)
print(splitter_docs[1].page_content)
print(len(splitter_docs[2].page_content))
print(len(splitter_docs[3].page_content))
print(splitter_docs[75].page_content)
print(len(splitter_docs))
```

```python
from sentence_transformers import SentenceTransformer

embedding_model = SentenceTransformer('all-MiniLM-L6-V2')

# Extract the page_content from each Document object
texts_to_encode = [doc.page_content for doc in splitter_docs]

embeddings = embedding_model.encode(texts_to_encode)
```

```python
print(f"{len(embeddings),len(embeddings[0])}")
```

storing embedding in a vector store

```python
vector_db = FAISS.from_embeddings(
    text_embeddings=list(zip(texts_to_encode, embeddings)),
    embedding=embedding_model # Although not used directly for embedding here, it's good practice to keep a reference to the model if needed later.
)
```

```python
query_embedding = embedding_model.encode("what is NLP")
similarity_docs  = vector_db.similarity_search_by_vector(query_embedding,k=2)
print(similarity_docs[0].page_content)
print()
print()
print(similarity_docs[1].page_content)
```

```python
from sentence_transformers import SentenceTransformer
from langchain.text_splitter import RecursiveCharacterTextSplitter

model = SentenceTransformer('all-MiniLM-L6-V2')

long_doc = """France is a country in Europe. The Eiffel Tower is in Paris. Paris is the capital city.
It is known for its cuisine and art. The Louvre Museum is a popular attraction. Python is also taught in many schools in France."""

text_splitter =  RecursiveCharacterTextSplitter(
    chunk_size = 50 ,chunk_overlap = 10
)
chunks =  text_splitter.split_text(long_doc)

embedding =  model.encode(chunks,convert_to_numpy=True)




```

#INDEDING USING FAISS VECTORDATABASE

```python
import faiss
import numpy as np

embedding_dim = embedding.shape[1]
index =  faiss.IndexFlatL2(embedding_dim) # we can use flatIP for consine similarity

index.add(embedding)
```

```python
print("Total vectors indexed:", index.ntotal)
store_vector =  index.reconstruct(0)
print("vector at index 0:",store_vector)

import numpy as np

original_vector = embedding[0]
print("Is stored vector same as original?", np.allclose(original_vector, store_vector))

```
